```python
0.#Vamos a crear un vectorfield con las ecuaciones de Van der Pol, basándonos en un código para las ecuaciones de Lotka-Volterra
from numpy import *
from scipy.integrate import odeint 

def dX_dt(w, t=0, p=0):
    
    x, y = w
    
    
    """ Return the ecuation for the Van der Pol oscillator.  """
    f = [ y,  p * (1 - x ** 2) * y - x]
    
    return f
```

```python
#Para la gráfica de fase, vamosa  crear nuevos documentos con diferentes condiciones inicales
#para llamarlas todas en una misma celda de gráfica.

#cambio de variables 1
mu = 2.0

#cambio de soluciones iniciales
x = -3
y = -3

# Parametros de la ED
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 170.0
numpoints = 5000

t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

#Ponemos las variables del vector:
w0 = [x, y]

# Llamamos a la funcion para resolver la ED
wsol = odeint(dX_dt, w0, t, args=(mu,),
              atol=abserr, rtol=relerr)

with open('Act8.2.dat', 'w') as f:
    # Imprimimos en el documento la solución
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```

```python
#1python
X_f0 = array ([ 1.0 , 1.0])
X_f1 = array ([ 0.0 , 0.0])
```

```python
from numpy import loadtxt
from pylab import figure, plot, xlabel, xlim, ylim, ylabel, grid, hold, legend, title, savefig
from matplotlib.font_manager import FontProperties
from scipy import integrate
%matplotlib inline

#Var. 1
t, x, y = loadtxt('Act8.2.dat', unpack=True)
plot(x, y, 'b-')

#Var.2
t, x, y = loadtxt('Act8.3.dat', unpack=True)
plot(x, y, 'b-')

#Var. 3
t, x, y = loadtxt('Act8.4.dat', unpack=True)
plot(x, y, 'b-')

#Var. 4
t, x, y = loadtxt('Act8.5.dat', unpack=True)
plot(x, y, 'b-')









import pylab as p


values  = linspace(6.0, 6.0, 5)                          # position of X0 between X_f0 and X_f1
vcolors = p.cm.autumn_r(linspace(0.3, 1., len(values)))  # colors for each trajectory

f2 = p.figure()

#-------------------------------------------------------
# plot trajectories
for v, col in zip(values, vcolors):
    X0 = v * X_f1                               # starting point
    X = integrate.odeint( dX_dt, X0, t)         # we don't need infodict here
    p.plot( X[:,0], X[:,1], lw=3.5*v, color=col, label='X0=(%.f, %.f)' % ( X0[0], X0[1]) )

#-------------------------------------------------------
# define a grid and compute direction at each point
nb_points   = 20

x1 = linspace(-6.0, 6.0, nb_points)
y1 = linspace(-6.0, 6.0, nb_points)

X1 , Y1  = meshgrid(x1, y1)                       # create a grid
DX1, DY1 = dX_dt([X1, Y1])                      # compute growth rate on the gridt
M = (hypot(DX1, DY1))                           # Norm of the growth rate 
M[ M == 0] = 1.                                 # Avoid zero division errors 
DX1 /= M                                        # Normalize each arrows
DY1 /= M










#Ahora voy a crear la gráfica de fase cuando está estable stalandome los primeros 100 renglones
t, x, y = loadtxt('Act8.5.dat', unpack=True, skiprows = 100)
plot(x, y, 'r-')
Q = p.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=p.cm.jet)
grid()
xlim(-6,6)
ylim(-6, 6)
xlabel('x')
ylabel('y')
title('Phase portrait of the unforced Van der Pol oscillator, showing a limit cycle')
f2.savefig('Image1.png')
```
