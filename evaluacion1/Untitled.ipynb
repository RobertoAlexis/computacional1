```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Leer archivo de datos
# Convertir las columnas Temp, de objeto a n√∫mero
df = pd.read_csv("sargento-salinidad-2702181.csv", header=None,skiprows=2, names=['#','DateTime','CondHigh','Temp','SpecificConductance','Salinity'])
#df.DateTime=pd.to_numeric(df.DateTime, errors='coerce')
df.CondHigh=pd.to_numeric(df.CondHigh, errors='coerce')
df.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df['Ndate'] = pd.to_datetime(df['DateTime'], format='%m/%d/%Y %H:%M:%S')
df['month'] = df['Ndate'].dt.month
df.head()
```

```python
# graficar Boxplots por mes
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="Salinity", data=df)
plt.show()
```

```python
# graficar Boxplots por mes
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="Temp", data=df)
plt.show()
```

```python
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("Temp", "Salinity", data=df, kind="reg",
                   color="r", size=7)
plt.show(g)
```
